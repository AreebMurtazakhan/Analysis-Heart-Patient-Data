### Converting Heart Disease Col Into Numeric

```python
dataset["HeartDisease"] = dataset["HeartDisease"].replace(0,"Normal").replace(1,"Heart Disease")
```

```python
dataset["HeartDisease"]
```

```python
dataset["FastingBS"] = dataset["FastingBS"].replace(0 , "Low").replace(1 , "High")
```

```python
# Separate The Categorical Columns
categorical_features = list(dataset.select_dtypes("object").columns)
print(f"Categorical columns : {categorical_features} \n")

# Separate The Numeric Columns
numeric_features = list(dataset.select_dtypes(["float64","int64"]).columns)
print(f"Numerical columns : {numeric_features} ")
```

### Bar Chart that displays the number of rows for each category of that column

```python
label = ["Sex" , "Chest Pain Type" , "Fasting Blood Sugar"\
         , "Resting Electrocardiogram" , "Exercise-Induced Angina"\
         , "ST_Slope" , "Heart Disease"]
```

```python
plt.figure(figsize=(14,12))
for i,ind,lab in zip(categorical_features,range(1,8) , label):
    count = dataset[i].value_counts()
    plt.subplot(4,2,ind)
    sns.barplot(count.index , count.values )
    plt.title(lab)
plt.tight_layout()
plt.show()
```

```python
for i in categorical_features:
    print(dataset[i].value_counts(),"\n")
```

####  Create a bar chart that displays the number of rows for each category of that column, grouped by HeartDisease.

```python
plt.figure(figsize=(16,10))
for i,ind,lab in zip(categorical_features[:6],range(1,7),label[:6]):
    group = dataset[categorical_features].groupby(["HeartDisease",i]).size().reset_index(name="Count")
    
    plt.subplot(3,2,ind)
    sns.barplot(data=group, x=i, y='Count', hue='HeartDisease')
    plt.title(f'Number of Rows for Each {lab} Grouped by Heart Disease\n')
    plt.xlabel(lab)
    plt.ylabel('Count')

plt.tight_layout()    
plt.show()
```

### Percentage of Patients with Heart Disease are Male and over 50 years old

```python
heart_disease = dataset[dataset["HeartDisease"] == "Heart Disease"]
male_50 = heart_disease[(heart_disease["Sex"] == "M") & (heart_disease["Age"] > 50)]
per_male_50 = ( len(male_50) / len(heart_disease) ) * 100 
per_male_50
```

```python
labels = ['Male over 50', 'Other']
sizes = [per_male_50, 100 - per_male_50]

plt.figure(figsize=(15,7))
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Percentage of Male Patients over 50 with Heart Disease \n' , fontsize = 15)
plt.show()
```

####  Replace those values with the median value of the corresponding column and the corresponding HeartDisease value

```python
plt.figure(figsize=(16,8))
plt.subplot(1,2,1)
plt.hist(dataset["RestingBP"])
plt.title("Resting Blood Pressure" , fontsize = 15)

plt.subplot(1,2,2)
plt.hist(dataset["Cholesterol"])
plt.title("Serum Colesterol" , fontsize = 15)

plt.show()
```

##### Observing the Two Bar Charts:
    * Resting Blood Pressure Shows the Left Skewed
    * Serum Cholesterol Shows the Right Skewed
###### So We the value of zero into median"

```python
dataset["Cholesterol"].median()
```

```python
dataset["Cholesterol"].replace(0,dataset["Cholesterol"].median(),inplace=True)
```

```python
dataset["RestingBP"].median()
```

```python
dataset["RestingBP"].replace(0,dataset["RestingBP"].median(),inplace=True)
```

```python
plt.figure(figsize=(16,8))
plt.subplot(1,2,1)
plt.hist(dataset["RestingBP"])
plt.title("Resting Blood Pressure" , fontsize = 15)

plt.subplot(1,2,2)
plt.hist(dataset["Cholesterol"])
plt.title("Serum Colesterol", fontsize = 15 )

plt.show()
```

### Convert the categorical features into dummy variables.

```python
dummy = pd.get_dummies(dataset[categorical_features] , drop_first=True)
dummy
```

### Pearson's correlation heat map.

```python
correl = dummy.corr()
correl
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(correl, annot=True, vmin=-1, vmax=1)
plt.title("Pearson's Correlation Heatmap \n" , fontsize = 15 )
plt.show()
```

###### Analysis
    * Chest Pain Type (ATA) is  Weak Positive Correlation With ST_Slope(Up) : (0.36) 
    * Chest Pain Type (NAP) is Weak Positive Correlation With Heart Disease(Normal) : (0.21)
    * Fast Blood Pressure (Low) is Weak Positive Correlation With Heart Disease(Normal) : (0.27)
    * Exercise Angina (Yes) is Weak Positive Correlation With ST_Slope(Flat) : (0.38)
    * ST_Slope (Up) is Moderate Positive Correlation with Heart Disease(Normal) : (0.62)
